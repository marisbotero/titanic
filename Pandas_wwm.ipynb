### 救Merge and join

Ambas funciones permiten que los datos de diferentes dataframes se combinen en uno solo de acuerdo con una regla de "cruce" o "b칰squeda".

Aunque tanto `merge` como` join` hacen cosas similares, la forma en que lo hacen es diferente.

La funci칩n `merge` es la funci칩n predeterminada de pandas para unir datos. B치sicamente es contraparte de *pandas de la uni칩n de SQL*, y requiere la especificaci칩n de qu칠 columnas de ambos dataframes se comparar치n. A Merge no le importa en absoluto los 칤ndices definidos en ellos.

Por otro lado, la funci칩n `join` de Panda es m치s conveniente (incluso utiliza merge internamente), unir es b치sicamente hacer una fusi칩n aprovechando los 칤ndices de ambos marcos de datos.

La siguiente figura resume los diferentes 4 tipos de combinaciones: _inner, outer, left and right_.

![MERGE](http://www.datasciencemadesimple.com/wp-content/uploads/2017/09/join-or-merge-in-python-pandas-1.png)

La funci칩n merge tambi칠n est치 disponible como m칠todo en la clase `DataFrame`.
La sintaxis b치sica es:

```
new_joined_df = df.merge (another_df, left_on = "col_in_df", right_on = "col_in_another_df",
먝먝먝먝먝먝먝먝먝먝먝먝먝먝먝먝먝먝먝먝먝먝먝먝먝맏ow="inner"|"left"|"right"|"outer")
```

El primer argumento (`another_df`), as칤 como` left_on` y `right_on` son argumentos obligatorios.
`left_on` especifica un nombre de columna en el dataframe `df` cuyos valores deben coincidir con
los de la columna `another_df` 'especificados en `right_on`.

El argumento `how` es opcional y por defecto es `inner`.

````
staff_df = pd.DataFrame([{'Name': 'Kelly', 'Role': 'Director of HR'},
                         {'Name': 'Sally', 'Role': 'Course liasion'},
                         {'Name': 'James', 'Role': 'Grader'}])
staff_df = staff_df.set_index('Name')
student_df = pd.DataFrame([{'Name': 'James', 'School': 'Business'},
                           {'Name': 'Mike', 'School': 'Law'},
                           {'Name': 'Sally', 'School': 'Engineering'}])
student_df = student_df.set_index('Name')
````

````
staff_df
````

````
student_df
````

````
pd.merge(staff_df, student_df,left_index=True, right_index=True)
````

````
pd.merge(staff_df, student_df, how='outer', left_index=True, right_index=True)
````

````
pd.merge(staff_df, student_df, how='left', left_index=True, right_index=True)
````

````
staff_df = staff_df.reset_index()
student_df = student_df.reset_index()
````

````
pd.merge(staff_df, student_df, how="left",on="Name")
````

### 救ㄴgregaci칩n y Agrupamiento
Un an치lisis esencial de grandes datos es un resumen eficiente: agregaciones inform치ticas como ``sum()``, ``mean()``, ``median()``, ``min()`` y ``max()``, en el que un solo n칰mero da una idea de la naturaleza de un conjunto de datos potencialmente grande.
